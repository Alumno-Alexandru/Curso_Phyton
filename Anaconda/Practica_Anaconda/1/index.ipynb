```python
def validar_email(email):
    patron = r'^[\w\.-]+@[\w\.-]+\.\w+$'
    if re.match(patron, str(email)):
        return email
    else:
        return np.nan  # Dejar en NaN si no cumple

df['email'] = df['email'].apply(validar_email)
```

```python
df.sort_values(by="nombre")
```

```python
df = df.dropna(subset=['email'])
```

```python
import pandas as pd
```

```python
import numpy as np
```

```python
df = pd.read_csv("personas_sucio.csv")
```

```python
df['nombre'] = df['nombre'].str.title()
```

```python
df['edad'] = pd.to_numeric(df['edad'], errors='coerce')
```

```python
df['edad'] = pd.to_numeric(df['edad'], errors='coerce')
media_edad = df['edad'].replace(0, np.nan).mean()
df['edad'] = df['edad'].replace(0, np.nan).fillna(media_edad).astype(int)
```

```python
df['ciudad'] = df['ciudad'].str.upper()
```

```python
df['email'] = df['email'].astype(str).str.strip()
patron_email = r'^[A-Za-z0-9._%+-]+@[A-Za-z0-9.-]+\.[A-Za-z]{2,}$'
df.loc[~df['email'].str.match(patron_email, na=False), 'email'] = 'desconocido@correo.com'
```

```python
df['fecha_registro'] = pd.to_datetime(df['fecha_registro'], errors='coerce', dayfirst=True)
df['fecha_registro'] = df['fecha_registro'].fillna(pd.Timestamp.now())
df['fecha_registro'] = df['fecha_registro'].dt.strftime('%d/%m/%Y')
```

```python
df.to_csv("datos_limpios.csv", index=False)
```
